```python
import pandas as pd
import numpy as np
```

```python
output_df = pd.read_csv('higarg_out.csv',delimiter='\t')
output_df.head(5)
```

```python
output_df[['1','2','g_truth']] = output_df['seq_id'].str.split('|',n=2, expand=True)
output_df.head(5)
```

```python
output_df['pred'] = output_df['pred'].str.split(';')
output_df['pred'] = output_df['pred'].fillna('[nonARG]')
output_df['g_truth'] = output_df['g_truth'].str.split(';')
```

```python
output_df.head(15)
```

```python
output_df.columns.to_list()[3:-3]
```

```python
classes = output_df.columns.to_list()[3:-3]
classes.append('nonARG')
output_df.head(15)
```

```python
len(classes)
```

```python

# mapping ground truth labels to the unique classes 
output_df['g_truth_update'] = output_df.g_truth.apply(lambda x: np.append(x,'others') if len(x) >=2 else x) 
g_values = output_df['g_truth_update'].values
intersection = []
for i in range(len(g_values)):
    intersection.append(np.intersect1d(g_values[i],classes))
output_df['g_truth_updated'] = intersection


list_nonarg = ['nonARG']
output_df['pred_modified'] = output_df.pred.apply(lambda x: list_nonarg if x == '[nonARG]' else x)  
```

```python
output_df = output_df[['g_truth_updated', 'pred_modified']]
```

```python
#output_df.where(output_df.g_truth == 'nonARG')
output_df['intersection_length'] = [len(set(a) & set(b)) for a, b in zip(output_df.g_truth_updated, output_df.pred_modified)]
output_df['intersection'] = [set(a) & set(b) for a, b in zip(output_df.g_truth_updated, output_df.pred_modified)]
output_df['union_length'] = [len(set(a) | set(b)) for a, b in zip(output_df.g_truth_updated, output_df.pred_modified)]
output_df['pred_length'] = output_df['pred_modified'].str.len()
output_df['gt_length'] = output_df['g_truth_updated'].str.len()
```

```python
output_df['g_truth_new'] = output_df['g_truth_updated'].apply(lambda x: np.append(x,'others') if len(x)==0 else x)
```

```python
output_df['gt_length'] = output_df['g_truth_new'].str.len()
num_samples = len(output_df)
precision = sum(list(output_df['intersection_length']/output_df['pred_length']))/num_samples
recall = sum(list(output_df['intersection_length']/output_df['gt_length']))/num_samples
f_score = 2*sum(list(output_df['intersection_length']/(output_df['pred_length'] + output_df['gt_length'])))/num_samples
accuracy =sum(list(output_df['intersection_length']/output_df['union_length']))/num_samples
```

```python
precision, recall, f_score, accuracy
```

```python
import sklearn.metrics
```

```python
y_true = output_df['g_truth_new'].to_list()
y_pred = output_df['pred_modified'].to_list()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import multilabel_confusion_matrix
```

```python
mlb = MultiLabelBinarizer()
y_true = mlb.fit_transform(y_true)
y_pred = mlb.fit_transform(y_pred)
```

```python
print('Precision: {0}'.format(sklearn.metrics.recall_score(y_true=y_true, y_pred=y_pred, average='micro')))
print('Recall: {0}'.format(sklearn.metrics.precision_score(y_true=y_true, y_pred=y_pred, average='micro'))) 
print('F1 Measure: {0}'.format(sklearn.metrics.f1_score(y_true=y_true, y_pred=y_pred, average='micro'))) 
```

```python
import math
```

```python
mcc = 0

for matrix in multilabel_confusion_matrix(y_true, y_pred):
    tn, fp, fn, tp = matrix.ravel()
    mcc += ((tp*tn - fp*fn)/math.sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn)))

mcc /= (multilabel_confusion_matrix(y_true, y_pred)).shape[0]
```
